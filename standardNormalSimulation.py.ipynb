```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
expectedValue, stdDeviation, numberOfSamples = 72,100,12
```

```python
randomTemperatureValues = np.random.normal(expectedValue,stdDeviation, numberOfSamples)
```

```python
temperatureList = list(randomTemperatureValues)
```

```python
plt.plot(temperatureList)
```

```python
temperatureList.sort()
```

```python
weatherValues = [None] * len(temperatureList)
```

```python
j=0
for i in range(0,len(temperatureList)):
    if i % 2 == 0:
        weatherValues[j] = temperatureList[i]
        j = j + 1
    else:
        weatherValues[ len(temperatureList) - j ] = temperatureList[i]
```

```python
plt.plot(weatherValues)
```
