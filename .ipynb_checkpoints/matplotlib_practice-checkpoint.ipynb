```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis , y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(x_axis , y_axis , marker="*", color = "green" , linewidth=2)

# Add a legend
ax.legend("Boston")

# Change the marker to a diamond

# Add a title and axes label
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel("Fare($)")

# Add a y-axis limit
ax.set_ylimt(0,45)
ax.set_xlimit(0,11)
# Add grid lines
ax.grid()
```

```python
# Create a scatter plot using the MATHLAB method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# Change the color of markers to red
plt.scatter(y_axis, x_axis, c= 'red')
# Add a legend for the city of Chicago
plt.legend('Chicago')

# Add a title and axes labels
plt.title ("PyBer Fare by Month")

# Switch the axis so that the Fare ($) data is on the x- axis
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.xlim(0, 50)

# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s= y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a bubble char using the object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#Change the color of the markers to sky blue.
plt.scatter(y_axis, x_axis, color = "skyblue", edgecolors = 'black', s = [ i * 5 for i in y_axis])

#Make the color 20% transparent.

#Add a legend for the city of Boston.
plt.legend('Boston')

#Add a title
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')

plt.xlim(0, 50)

#Invert the y-axis so that January is at the top. 
plt.gca().invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.subplots(figsize=(8,8))
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
```
